- cuts recordings down to 2.88sec to cut off long tails

```python
from pydub import AudioSegment
import os
import re

# where the unformatted audio is stored
audio_folder = r"C:/Users/bw339/OneDrive - University College London/Mars internship/Bomb classifier/data/processed_n_bombs_10k/"

# where the new audio will be output
output_folder = r"C:/Users/bw339/OneDrive - University College London/Mars internship/Bomb classifier/data/processed_n_bombs_10k1/"
```

OPTIONAL

Add leading zeroes to file names moth number if single digit, e.g M5 becomes M05

```python
directory = audio_folder

# Iterate over all files in the directory
for filename in os.listdir(directory):
    # Check if the file name matches the expected format
    if re.match(r"NB\d+_[Mm](\d)_\d{8}_\d{6}", filename):
        # Extract the number after the "M"
        m_number = re.search(r"[Mm](\d)", filename).group(1)
        # If the number is below 10, add a leading zero
        if int(m_number) < 10:
            new_filename = re.sub(r"(?<=[Mm])\d", "0" + m_number, filename)
            # Rename the file
            os.rename(os.path.join(directory, filename), os.path.join(directory, new_filename))
            print(f"Renamed {filename} to {new_filename}")
```

MAIN LOOP

Downsamples to 8kHz (so 4kHz audible) and cut to 2.88sec in length

```python
# main loop
counter = 0
for filename in os.listdir(audio_folder):
    if filename.endswith(".wav") or filename.endswith(".WAV"):
        try:
            counter += 1
            # Open the file and read its content
            sound = AudioSegment.from_file(f"{audio_folder}/{filename}", format="wav")

            # Downsample the audio to 8kHz
           # sound = sound.set_frame_rate(8000)  ################ why commented out???????
            
            # Cut the audio to 2.88 seconds
            sound = sound[:2880]
            
            # Write the new audio to a file in folder2
            sound.export(f"{output_folder}/{filename}", format="wav")
            print('Exported :' + filename)
        except:
            print('Error at file (skipping): ' + filename)
            continue

# count number of files in output folder
print(f'Made {counter} new files')
```
